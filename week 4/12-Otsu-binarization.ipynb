# Applying Thresholding with Automatic Threshold Selection using Otsu’s Method**

### 🔹 Description

Otsu’s method **automatically calculates the optimal threshold** for converting a grayscale image into a binary image.  
It is especially useful when the image has **a bimodal histogram** (two distinct intensity peaks), such as:

- Foreground (object)  
- Background  

This method eliminates the need to manually choose the threshold value.

---

### 🔹 Python / OpenCV Implementation

```python
import cv2
from matplotlib import pyplot as plt

# Load the grayscale image
image = cv2.imread("car_plate.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Otsu's thresholding
# Set threshold=0 and add cv2.THRESH_OTSU flag
_, binary_otsu = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the result
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(binary_otsu, cmap='gray')
plt.title("Otsu Thresholding")
plt.axis('off')

plt.show()


````python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread("C:/Users/sudob/OneDrive/Desktop/week4/images.jpg",0)
# global thresholding
ret1,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
# Otsu's thresholding
ret2,th2 = cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
# Otsu's thresholding after Gaussian filtering
blur = cv.GaussianBlur(img,(5,5),0)
ret3,th3 = cv.threshold(blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

# plot all the images and their histograms
images = [img, 0, th1,
          img, 0, th2,
          blur, 0, th3]
titles = ['Original Noisy Image','Histogram','Global Thresholding (v=127)',
          'Original Noisy Image','Histogram',"Otsu's Thresholding",
          'Gaussian filtered Image','Histogram',"Otsu's Thresholding"]

plt.figure(figsize=[15,7])
for i in range(3):
    #first column
    plt.subplot(3,3,i*3+1),plt.imshow(images[i*3],'gray'),
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
    #second column
    plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
    plt.title(titles[i*3+1]), plt.xticks([]), plt.yticks([])
    #third column
    plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],'gray')
    plt.title(titles[i*3+2]), plt.xticks([]), plt.yticks([])
plt.show()
````

# Applying Automatic Thresholding on a License Plate Image**

### 🔹 Description

Automatic thresholding (Otsu’s method) can be applied to a **grayscale image of a car license plate** to:

- Automatically determine the best threshold value.
- Convert the image into a **binary form** where characters on the plate are clearly highlighted.
- Simplify further processing such as **OCR (Optical Character Recognition)** or license plate recognition.

---

### 🔹 Python / OpenCV Example

```python
import cv2
from matplotlib import pyplot as plt

# Load the license plate image in grayscale
plate_image = cv2.imread("car_plate.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Otsu's automatic thresholding
_, binary_plate = cv2.threshold(
    plate_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# Display the original and binary images
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(plate_image, cmap='gray')
plt.title("Original License Plate")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(binary_plate, cmap='gray')
plt.title("Binary Image (Otsu Threshold)")
plt.axis('off')

plt.show()

# Applying Automatic Thresholding on a Dark License Plate Image**

### 🔹 Description

When a license plate image is **dark or poorly illuminated**, automatic thresholding (Otsu’s method) can:

- Automatically determine the optimal threshold to separate the characters from the dark background.
- Convert the image into a **binary format**, making the letters or numbers clearly visible.
- Facilitate further processing such as **OCR** or license plate recognition.

---

### 🔹 Python / OpenCV Example

```python
import cv2
from matplotlib import pyplot as plt

# Load the dark license plate image in grayscale
dark_plate = cv2.imread("dark_car_plate.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Otsu's automatic thresholding
_, binary_dark_plate = cv2.threshold(
    dark_plate, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# Display results
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(dark_plate, cmap='gray')
plt.title("Original Dark License Plate")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(binary_dark_plate, cmap='gray')
plt.title("Binary Image (Otsu Threshold)")
plt.axis('off')

plt.show()


##  Key Points of Otsu’s Thresholding for Dark License Plates

- **Automatic Threshold Calculation:**  
  Otsu’s method computes the optimal threshold value automatically, even for images with **low brightness** or poor lighting.

- **Binary Conversion:**  
  Transforms dark license plate images into a **clear binary image**, where the foreground (characters) and background are distinctly separated.

- **Character Highlighting:**  
  The characters on the plate are emphasized, making **detection and recognition easier** for further processing such as OCR.

- **Enhancement Compatibility:**  
  Can be combined with **contrast enhancement** or **morphological operations** (erosion, dilation, opening, closing) to improve visibility and segmentation of characters in very dark images.
